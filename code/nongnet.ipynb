```python
import pandas as pd
import urllib.request
import json
import datetime
import os
```

```python
### 어제의 농넷 데이터 불러오기
### 시간별, 도매시장별로 데이터가 계속 추가 업데이트 됩니다.
### 전체 데이터를 얻기 위해서 하루 전날(D-1)을 선택.
target_start = datetime.date(2021, 8, 24)
target_end = datetime.date(2021, 9, 27)
target_path = '../data/private/tsalet'
input_window = 112
target_window = 28
url = 'https://www.nongnet.or.kr/api/whlslDstrQr.do?sdate=' # sdate = 날짜
```

```python

st = target_start - datetime.timedelta(days=input_window)
for i in range((target_end - target_start).days + input_window + target_window + 1):
    delta = datetime.timedelta(days=i)
    target = (st + delta).strftime('%Y%m%d')
    if os.path.exists(os.path.join(target_path, '{}.csv'.format(target))):
        continue

    response = urllib.request.urlopen(url+target).read()
    response = json.loads(response)

    data = pd.DataFrame(response['data']) # ALL_TSALET_ALL_ 데이터와 같은 형식
    data.to_csv(os.path.join(target_path, '{}.csv'.format(target)))
    print('Saved :', target)
print("Done")
```
