AT_TSALET_ALL파일로부터 train.csv 파일 생성하는 코드

대상 외 품목/품종 추가, 기타 column 등 추가하여 feature로 활용 가능

```python
import pandas as pd
import numpy as np
from glob import glob
import multiprocessing
import pickle

import preprocessing
```

```python
name = 'private'
```

```python
tsalet_files = sorted(glob('../data/private/2014/*.csv'))
pool = multiprocessing.Pool(processes=multiprocessing.cpu_count())
pool.map(preprocessing.preprocessing, tsalet_files)
pool.close()
pool.join()
```

```python
dict_files = sorted(glob('../data/'+name+'/dict/*.pkl'))
```

```python
train_dict_list = []
for dict_file in dict_files:
    with open(dict_file, 'rb') as f:
        train_dict = pickle.load(f)
    train_dict_list.append(train_dict)
```

```python
train = None

for train_dict in train_dict_list:
    if train is None:
        train = pd.DataFrame(train_dict)
    else:
        train = pd.concat([train, pd.DataFrame(train_dict)])
```

```python
train['date'] = train.date.astype(str).str.replace('-','')
train['date'] = pd.to_datetime(train.date, format='%Y%m%d')
```

```python
train
```

```python
t = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
train['요일'] = train['date'].apply(lambda x : t[x.weekday()])
order = list(train.columns)
del order[-1]
order.insert(1, '요일')
train = train[order]
train
```

```python
train.to_csv('../data/'+name+'_train.csv', index=False)
```
